```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from sklearn import linear_model
import math
```

```python
data= pd.read_csv('./marketing_customer_analysis.csv')
data.head()
```

```python
data.isna().sum()
```

```python
data= data.drop_duplicates()
data.shape
```

```python
#separate the features(we are trying to predict the Traget so this one will be the feature) from the labels
y= data['Total Claim Amount'] # this one is the feature we are trying to predict
X= data.drop(['Total Claim Amount'],axis=1) # this are the rest of labelss
```

```python
X_num= X.select_dtypes(include=np.number)
X_cat= X.select_dtypes(include=np.object)
```

```python
X_num.head()
```

```python
X_cat.head()
```

```python
from sklearn.preprocessing import StandardScaler
Standardtransformer = StandardScaler().fit(X_num)
X_standardized = Standardtransformer.transform(X_num)
print(X_standardized.shape)
X_standardized = pd.DataFrame(X_standardized,columns=X_num.columns)

```

```python
#what about categoricals?
X_cat.head()
```

```python
X_cat['Gender'].value_counts()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder().fit(X_cat[['Gender']])
print(encoder.categories_)
encoded = encoder.transform(X_cat[['Gender']]).toarray()
print(encoded)
onehot_encoded = pd.DataFrame(encoded,columns=encoder.categories_)
onehot_encoded.head()
```

```python
X_cat['Vehicle Size'].value_counts()
```

```python
encoder2 = OneHotEncoder().fit(X_cat[['Vehicle Size']])
print(encoder2.categories_)
encoded2 = encoder2.transform(X_cat[['Vehicle Size']]).toarray()
print(encoded2)
onehot_encoded2 = pd.DataFrame(encoded2,columns=encoder2.categories_)
onehot_encoded2.head()
```

```python
onehot_encoded2 = onehot_encoded2.drop(['Small'],axis=1)
onehot_encoded2.head()
```

```python
X_cat['Marital Status'].value_counts()
```

```python
encoder3 = OneHotEncoder().fit(X_cat[['Marital Status']])
print(encoder3.categories_)
encoded3 = encoder3.transform(X_cat[['Marital Status']]).toarray()
print(encoded3)
onehot_encoded3 = pd.DataFrame(encoded3,columns=encoder3.categories_)
onehot_encoded3.head()
```

```python
onehot_encoded3 = onehot_encoded3.drop(['Divorced'],axis=1)
onehot_encoded3.head()
```

```python
encoder4 = OneHotEncoder().fit(X_cat[['EmploymentStatus']])
print(encoder4.categories_)
encoded4 = encoder4.transform(X_cat[['EmploymentStatus']]).toarray()
print(encoded4)
onehot_encoded4 = pd.DataFrame(encoded4,columns=encoder4.categories_)
onehot_encoded4.head()
```

```python
onehot_encoded4 = onehot_encoded4.drop(['Unemployed'],axis=1)
onehot_encoded4.head()
```

```python
#ahora juntamos los numerical con los categorical convertidos en numerical
X = pd.concat([X_standardized, onehot_encoded, onehot_encoded2, onehot_encoded3, onehot_encoded4], axis=1)
X.head()
```

```python
y = data['Total Claim Amount']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1000)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
X_train.head()
```

```python
X_test.head()
```

```python
model = linear_model.LinearRegression()
model = model.fit(X_train,y_train)
```

```python
model.predict(X_test)
```

```python
from sklearn.metrics import r2_score

r2_score(y_test,model.predict(X_test))
```

```python
predicted= model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
from sklearn.metrics import r2_score

predictions= model.predict(X_test)
r2_score(y_test, predictions)
```

```python
mse = mean_squared_error(y_test, predictions)
print(mse)
```

```python
rmse = math.sqrt(mse)
print(rmse)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mae = mean_absolute_error(y_test, predictions)
print('Using scikit-lean, the error is {}'.format(mae))
```
